```python
import numpy as np
import pandas as pd
te=pd.read_csv("test_processed.csv")
va=pd.read_csv("valid_processed.csv")
```

```python
va
```

```python
# sensitive
from textblob import TextBlob

for i in range(0, len(va)):
    txt=TextBlob(va['reviewComment'][i])
    va['reviewComment'][i]=txt.sentiment
    
for i in range(0, len(va)):
    if va['reviewComment'][i][1]>0:
        va['reviewComment'][i]=1
    else:
        va['reviewComment'][i]=-1
```

```python
# sensitive
from textblob import TextBlob

for i in range(0, len(te)):
    txt=TextBlob(te['reviewComment'][i])
    te['reviewComment'][i]=txt.sentiment
    
for i in range(0, len(te)):
    if te['reviewComment'][i][1]>0:
        te['reviewComment'][i]=1
    else:
        te['reviewComment'][i]=-1
```
